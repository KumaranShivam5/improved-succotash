```
import numpy as np
import matplotlib.pyplot as plt
```

```
mu=[1,2]
sigma=[[12,8],[8,24]]
x=np.random.multivariate_normal(mu,sigma,1000).T
```

```
plt.plot(x[0],x[1],'+')
plt.axis('equal')
plt.show()
```

**PDF plot**

```
plt.hist2d(x[0],x[1],bins=80,density=False,cmap=plt.cm.jet)
plt.colorbar()
plt.show()
```

**DECORELATION**

```
eig_val,eig_vec=np.linalg.eig(sigma)
print(eig_val)
print(eig_vec)
y=np.dot(eig_vec.T,x)
print(y)

plt.plot(y[0],y[1],'+')
plt.axis('equal')
plt.show()
```

**WHITENING**

```
w=np.dot(np.linalg.inv(np.sqrt(eig_val)),y)
print(y)
print(w)

plt.plot(w[0],w[1],'+')
plt.axis('equal')
plt.show()
```

```
plt.hist2d(w[0],w[1],bins=80,density=False,cmap=plt.cm.jet)
plt.colorbar()
plt.show()
```
