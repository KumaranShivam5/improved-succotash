```
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```
def generate_data(mu_1,sigma_1,mu_2,sigma_2,p_w1,p_w2):
    x_w1=np.random.normal(mu_1,sigma_1,1000)
    x_w2=np.random.normal(mu_2,sigma_2,1000)

    bins = np.linspace(-10, 10, 40)
    hist_w1, bins = np.histogram(x_w1, bins=bins, density=True)
    hist_w2, bins = np.histogram(x_w2, bins=bins, density=True)
    bin_centers = 0.5*(bins[1:] + bins[:-1])
    pdf_w1 = stats.norm(mu_1,sigma_1).pdf(bin_centers)
    pdf_w2 = stats.norm(mu_2,sigma_2).pdf(bin_centers)
    
    


    #plt.figure(figsize=(6, 4))
    #plt.plot(bin_centers, pdf_w1, label="PDF")
    #plt.plot(bin_centers, pdf_w2, label="PDF")
    #plt.legend()
    #plt.show()

    calc_post(p_w1,p_w2,hist_w1,hist_w2,bin_centers)
```
